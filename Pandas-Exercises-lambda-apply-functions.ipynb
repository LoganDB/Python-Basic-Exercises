# Student Alcohol Consumption

### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the datasets *student-mat* and *student-por* and merge them.

```python
df1=pd.read_csv("student-mat.csv")
df2=pd.read_csv("student-por.csv")
```

### Step 3. Assign it to a variable called df.

```python
df=pd.merge(left=df1,right=df2,how='outer',sort=True)
df.head()
```

### Step 4. For the purpose of this exercise slice the dataframe from 'school' until the 'guardian' column

```python
stud_alcoh = df.loc[: , "school":"guardian"]
stud_alcoh.head()
```

### Step 5. Create a lambda function that captalize strings.

```python
capitalize = lambda x: x.capitalize()
```

### Step 6. Capitalize both Mjob and Fjob

```python
stud_alcoh['Mjob'].apply(capitalize)
stud_alcoh['Fjob'].apply(capitalize)
```

### Step 7. Print the last elements of the data set.

```python
stud_alcoh.tail()
```

### Step 8. Did you notice the original dataframe is still lowercase? Why is that? Fix it and captalize Mjob and Fjob.

```python
stud_alcoh['Mjob'] = stud_alcoh['Mjob'].apply(capitalize)
stud_alcoh['Fjob'] = stud_alcoh['Fjob'].apply(capitalize)
stud_alcoh.tail()
```

### Step 9. Create a function called majority that return a boolean value to a new column called legal_drinker

```python
def majority(x):
    if x > 18:
        return True
    else:
        return False
```

```python
stud_alcoh['legal_drinker'] = stud_alcoh['age'].apply(majority)
stud_alcoh.head()
```

### Step 10. Multiply every number of the dataset by 10. 

```python
def times_10(x):
    if type(x) is int:
        return 10 * x
    return x
```
